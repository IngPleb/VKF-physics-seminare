## Task 1

Create graph of the trajectory of a horizontal ball thrown with the following parameters:
- Initial velocity: 10 m/s
- Height: 100 m
- g = 10 m/s^2

```python
g = 10
h = 100
v_0 = 10
```

```python
# Time of flight
T = math.sqrt(2 * h / g)

t = np.linspace(0, T, num=100)
t
```

## Task 3

Create a graph (s) of the trajectory(ies) of a projectile(s) thrown with these parameters:

- Initial velocity: 15 m/s
- g = 10 m/s^2
- angle_1 = 15°
- angle_2 = 30°
- angle_3 = 45°
- angle_4 = 60°
- angle_5 = 75°
- angle_6 = 90°

```python
angles = np.array([15, 30, 45, 60, 75, 90])
v_0 = 15
g = 10
```

```python
def calculate_trajectory(angle):
    v_x = v_0 * math.cos(math.radians(angle))
    v_y = v_0 * math.sin(math.radians(angle))

    T = 2 * v_y / g
    t = np.linspace(0, T, num=100)

    x = v_x * t
    y = v_y * t - 0.5 * g * t**2

    return x, y
```

```python
x_trajectories = []
y_trajectories = []

for angle in angles:
    x, y = calculate_trajectory(angle)
    x_trajectories.append(x)
    y_trajectories.append(y)
```

```python
plt.figure(figsize=(8, 5))
for i, angle in enumerate(angles):
    plt.plot(x_trajectories[i], y_trajectories[i], label=f"Trajektorie {angle}°", marker='o', linestyle='-')
plt.xlabel("Horizontální vzdálenost (m)")
plt.ylabel("Vertikální vzdálenost (m)")
plt.title("Vrh s počáteční rychlostí 15 m/s a různými úhly")
plt.legend()
plt.grid()
plt.axhline(0, color="black", linewidth=0.5)
plt.show()
```

#### Alternative solution using numpy arrays

```python
angles = np.array([15, 30, 45, 60, 75, 90])
v_0 = 15
g = 10
```

```python
v_x = v_0 * np.cos(np.radians(angles))
v_y = v_0 * np.sin(np.radians(angles))

T = 2 * (v_y / g)

t = np.linspace(0, T, 100)
t.shape
```

```python
v_x.shape
```

```python
x = t * v_x
y = v_y * t - 1/2 * g * t ** 2
x.shape, y.shape
```

```python
plt.figure(figsize=(8, 5))  # Set figure size

# Loop over each trajectory (each column corresponds to an angle)
for i in range(x.shape[1]):
    plt.plot(x[:, i], y[:, i], label=f"Úhel {angles[i]}°", marker="o")  # Plot trajectory

plt.xlabel("Horizontální vzdálenost (m)")
plt.ylabel("Vertikální vzdálenost (m)")
plt.title("Vrh s počáteční rychlostí 15 m/s a různými úhly")
plt.legend()
plt.axhline(0, color="black", linewidth=0.5)
plt.grid()
plt.show()
```

## Task 4

Create a graph of the trajectory of a projectile with the following parameters:

- Initial velocity: 15 m/s
- g = 10 m/s^2
- angle = 50°
- The influence of air resistance is 5 m/s^1 per second.

```python
# Constants
v0 = 15          # initial speed in m/s
g = 10           # gravitational acceleration in m/s^2
alpha_deg = 50   # launch angle in degrees
alpha = math.radians(alpha_deg)

# Drag is modeled as a constant deceleration of 5 m/s^2
# This means that the overall speed decreases as: v(t) = 15 - 5*t

# Analytical expressions for the positions:
def x(t):
    # x(t) = cos(alpha) * (15t - (5/2)t^2)
    return math.cos(alpha) * (v0 * t - 2.5 * t**2)

def y(t):
    # y(t) = sin(alpha) * (15t - (5/2)t^2) - 5t^2
    return math.sin(alpha) * (v0 * t - 2.5 * t**2) - 5 * t**2

# Time of flight:
# We set y(t) = 0 for t > 0:
#   sin(alpha)*(15t - 2.5t^2) - 5t^2 = 0
#   t * [15 sin(alpha) - t*(2.5 sin(alpha) + 5)] = 0
# Ignoring t = 0, we solve:
#   t = (15 sin(alpha)) / (2.5 sin(alpha) + 5)
t_flight = (v0 * math.sin(alpha)) / (2.5 * math.sin(alpha) + 5)
print(f"Time of flight: {t_flight:.3f} s")

# Time to reach maximum height:
# The vertical velocity is the derivative of y(t):
# v_y(t) = d/dt [y(t)] = sin(alpha)*(15 - 5t) - 10t
# Setting v_y(t) = 0 and solving for t gives:
#   15 sin(alpha) - t*(5 sin(alpha) + 10) = 0  =>  t_max = (15 sin(alpha))/(5 sin(alpha) + 10)
t_max = (v0 * math.sin(alpha)) / (5 * math.sin(alpha) + 10)
print(f"Time at maximum height: {t_max:.3f} s")

# Maximum height
y_max = y(t_max)
print(f"Maximum height: {y_max:.3f} m")

# Horizontal range (x-coordinate at time of flight)
x_range = x(t_flight)
print(f"Horizontal range: {x_range:.3f} m")

# Plot the trajectory
t_values = np.linspace(0, t_flight, 200)
x_values = [x(t) for t in t_values]
y_values = [y(t) for t in t_values]

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label="Trajektorie", marker='o', linestyle='-')
plt.xlabel("Horizontální vzdálenost (m)")
plt.ylabel("Vertikální vzdálenost (m)")
plt.title("Vrh s počáteční rychlostí 15 m/s, úhlem 50° a odporovou silou 5 m/s^2 (analytické řešení - konstantní odpor)")
plt.axhline(0, color="black", linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
