### Imports

```python
import numpy as np
```

### Defining the constants

```python
g = 9.81 # m/s^2
rho_water = 970 # kg/m^3
rho_ethanol = 789 # kg/m^3
rho_mercury = 13579.04 # kg/m^3
```

### Define the function for pressure

```python
# 0 â€“ 10 meter steps
steps = np.arange(11) # meters
steps
```

```python
def depth_pressure(rho, depth = steps):
    return rho * g * depth
```

## Calculate pressure

```python
water_pressure = depth_pressure(rho_water)
water_pressure
```

```python
ethanol_pressure = depth_pressure(rho_ethanol)
ethanol_pressure
```

```python
mercury_pressure = depth_pressure(rho_mercury)
mercury_pressure
```

### Plotting the results

```python
import matplotlib.pyplot as plt

# Create the figure and axis
plt.figure()

# Plot the pressure data for each fluid with distinct line styles and colors
plt.plot(steps, water_pressure, label="Water", linestyle='-', color='b', marker='o')
plt.plot(steps, ethanol_pressure, label="Ethanol", linestyle='--', color='g', marker='s')
plt.plot(steps, mercury_pressure, label="Mercury", linestyle='-.', color='r', marker='^')

# Add labels for the x and y axes
plt.xlabel('Depth (meters)')
plt.ylabel('Pressure (Pa)')

# Add a title to the plot
plt.title('Pressure vs Depth for Water, Ethanol, and Mercury')


# Add a legend to differentiate between the fluids
plt.legend()

# Add a grid to the plot
plt.grid(True)

# Show the plot
plt.show()
```
