```python
from ctypes import py_object

import numpy as np
```

### Constatns

```python
steps = np.arange(0, 20, 0.3)
rho_water = 997 # kg/m^3
g = 9.81 # m/s^2
```

### Shape definition

```python
cube_edge = 10 # m
sphere_radius = 9 # m
```

```python
def calc_buoyancy_cube(cube_edge_length, depth, fluid_rho = rho_water, gravitational_acceleration = g):
    cube_base = cube_edge_length**2
    
    # We take the min between depth adn cube_edge_length to ensure that when the cube is submerged more than its side it doesn't gain more buoyancy force
    volume = cube_base * np.minimum(depth, cube_edge_length)
    return volume * fluid_rho * gravitational_acceleration
```

```python
cube_forces = [calc_buoyancy_cube(cube_edge, x) for x in steps]
cube_forces
```

```python
from math import pi as PI

def calc_buoyancy_sphere(radius, depth, fluid_rho = rho_water, gravitational_acceleration = g):
    depth_submersion = np.minimum(2 * radius, depth)
    volume = ((PI * depth_submersion**2) / 3) * (3 * radius - depth_submersion)
    return volume * fluid_rho * gravitational_acceleration
```

```python
sphere_forces = calc_buoyancy_sphere(sphere_radius, steps)
sphere_forces
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,4))
# Plot red circles for points ("ro") and connect them with a line ("r-")
plt.plot(steps, cube_forces, "ro-")

# Extend the x-axis range (for example, extending 1 unit on each side)
plt.xlim(min(steps) - 0.4, max(steps) + 1)

plt.title("Buoyancy Force Of A Cube With A Side Length Of 10 m (in Water)")
plt.xlabel("Depth (m)")
plt.ylabel("Buoyancy Force (Pa)")

plt.show()
```

```python
plt.figure(figsize=(16,4))
# Plot red circles for points ("ro") and connect them with a line ("r-")
plt.plot(steps, sphere_forces, "ro-")

# Extend the x-axis range (for example, extending 1 unit on each side)
plt.xlim(min(steps) - 0.4, max(steps) + 1)

plt.title("Buoyancy Force Of A Sphere With A Radius of 9 m (in Water)")
plt.xlabel("Depth (m)")
plt.ylabel("Buoyancy Force (Pa)")

plt.show()
```
